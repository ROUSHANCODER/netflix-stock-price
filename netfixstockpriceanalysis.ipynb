```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objs as go
```

```python
df=pd.read_csv("NFLX.csv")
```

```python
df
```

```python
df.head()
```

```python
sns.set(rc={'figure.figsize':(10,5)})
```

```python
df['Date']=pd.to_datetime(df['Date'])
df=df.set_index('Date')
df.head()
```

```python
sns.lineplot(x=df.index,y=df['Volume'],label='Volume')
plt.title('volume f stock versus time')
```

```python
df.plot(y=['High','Close','Open'],title='NETFLIX STOCK PRICE')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, figsize=(15, 10))
# Plot mean volume grouped by day
df.groupby(df.index.day).mean().plot(y='Volume', ax=ax1, xlabel='Day')
ax1.set_title('Mean Volume by Day')

# Plot mean volume grouped by month
df.groupby(df.index.month).mean().plot(y='Volume', ax=ax2, xlabel='Month')
ax2.set_title('Mean Volume by Month')

# Plot mean volume grouped by year
df.groupby(df.index.year).mean().plot(y='Volume', ax=ax3, xlabel='Year')
ax3.set_title('Mean Volume by Year')
```
